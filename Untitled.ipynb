```python
import seaborn as sns

df_titanic = sns.load_dataset('titanic')
df_titanic.dropna(axis=0, inplace=True)
df_titanic.head()
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
from sklearn import neighbors, datasets
from sklearn import preprocessing


x_feat_list = ['age', 'fare']
y_feat = 'alive'

list_of_user_input = {'alive': 'Unknown'}
for item in x_feat_list:
    value_from_user = float(input(f'please enter the {item}'))
    list_of_user_input[item] = value_from_user
list_of_user_input

n_neighbors = 3

# prepare data
x = df_titanic.loc[:, x_feat_list].values
y_true = df_titanic.loc[:, y_feat].values


# we create an instance of Neighbours Classifier and fit the data.
clf = neighbors.KNeighborsClassifier(n_neighbors, weights='distance')
clf.fit(x, y_true)

list_of_data = []
for feat in x_feat_list:
    list_of_data.append(list_of_user_input[feat])
    
dataClass = clf.predict([list_of_data])
print(dataClass)
```
